````python
# Solution
def contar_duplicados(Google_Play_sin_10472):
    total_filas=len(Google_Play_sin_10472)
    filas_únicas=len(Google_Play_sin_10472.drop_duplicates())
    filas_duplicadas = total_filas - filas_únicas
    return filas_duplicadas 

duplicados = contar_duplicados(Google_Play_sin_10472)
print( f'{duplicados} valores duplicados' )
````

5. a. Create a dictionary, where each dictionary key is a unique app name and the corresponding dictionary value is the highest number of reviews of that app


````python
# Solution
def create_app_reviews_dictionary(dataframe):
    
    app_reviews_dict = {}

    for index, row in dataframe.iterrows():
        app_name = row['App']
        app_reviews = row['Reviews']

        if app_name not in app_reviews_dict or app_reviews > app_reviews_dict[app_name]:
            app_reviews_dict[app_name] = app_reviews

    return app_reviews_dict
result_dict = create_app_reviews_dictionary(Diccionario_sin_duplicados)
````

 8. Build a frequency table for the ```prime_genre``` column of the App Store dataset, and the ```Genres``` and ```Category``` columns of the Google Play data set. What are the most common genres for each market? You must develop and deploy your function

````python
# Solution

def frequency_table(dataframe, column_name):
    return dataframe[column_name].value_counts().reset_index().rename(columns={'index': column_name, column_name: 'Frequency'})

# Frecuencias de 'prime_genre'
app_store_frequency = frequency_table(apps_gratuitas2, 'prime_genre')

# Frecuencias de 'Genres'
google_play_genres_frequency = frequency_table(apps_gratuitas, 'Genres')

# Frecuencias de 'Category'
google_play_category_frequency = frequency_table(apps_gratuitas, 'Category')
print(app_store_frequency.head())
print(google_play_genres_frequency.head())
print(google_play_category_frequency.head())
````
